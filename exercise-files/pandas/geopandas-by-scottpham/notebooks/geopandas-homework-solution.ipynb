```python
import geopandas as gpd
import pandas as pd
```

```python
%matplotlib inline
```

```python
states = gpd.read_file("../data/states")
states.head()
```

```python
ufo_sightings = pd.read_csv("../data/ufo_sightings_data.csv")
ufo_sightings.head()
```

### Merge your data here

```python
ufo_data = states.merge(
    ufo_sightings,
    left_on="NAME",
    right_on="state"
)

ufo_data
```

### Plot your data here

```python
ax = ufo_data.plot(
    column = "sightings",
    legend = True
)
```

### Now export a chart

```python
# first, get the "Figure" from the plot object.
fig = ax.get_figure()

# then, use the savefig method on that Figure object
fig.savefig("../output/ufo_sightings.png")
```
